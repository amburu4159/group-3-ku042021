```python
import numpy as np
import pandas as pd
import os
import joblib
from joblib import dump, load
```

```python
#Reading in data from ML model cleaning process
df = pd.read_csv("MLData.csv")

#Reading in new data set
test = pd.read_csv("Prediction-Data-Set.csv")
```

```python
#Checking Test set data
test
```

```python
#Checking ML model data
df
```

```python
#Dropping columns
df = df.drop(columns=['Unnamed: 0', 'Number of Deaths'])
df
```

```python
#Adding new data onto set from ML
append_df = df.append(test, ignore_index=True)
append_df
```

```python
# Generate our categorical variable lists
categories = append_df.dtypes[append_df.dtypes == "object"].index.tolist()
categories
```

```python
from sklearn.preprocessing import OneHotEncoder

# Create a OneHotEncoder instance
enc = OneHotEncoder(sparse=False)

# Fit and transform the OneHotEncoder using the categorical variable list
encode_df = pd.DataFrame(enc.fit_transform(append_df[categories]))

# Add the encoded variable names to the dataframe
encode_df.columns = enc.get_feature_names(categories)
encode_df
```

```python
# Merge one-hot encoded features and drop the originals
merge_df = append_df.merge(encode_df, left_index=True, right_index=True)
merge_df = merge_df.drop(categories, 1)
merge_df
```

```python
list(merge_df.columns)
```

```python
#Set features and target variables
X = merge_df
```

```python
from sklearn.preprocessing import StandardScaler

#Fit scaler to the training set of features
scaler = StandardScaler().fit(X)

X_scaled = scaler.transform(X)
X_scaled
```

```python
#Checking size of array
X_scaled.shape
```

```python
#Printing out list of array beyond the last row of original machine learning data
testing_data = X_scaled[66057:]
testing_data
```

```python
#Loading in model
model = joblib.load('model.pkl')
```

```python
#Checking model
model
```

```python
#Making predictions based on the model
prediction = model.predict(testing_data)
```

```python
#Printing out predictions
prediction
```

```python
#Adding predictions column to test set dataframe
test['Death Predicted?'] = prediction
```

```python
#Checking dataframe
test
```

```python
#Using for loop to make all deaths in Number of Deaths say Yes-1 or No-0
for i in range(len(test)):
    if test["Death Predicted?"][i] == 0:
        test["Death Predicted?"][i] = "No"
    else:
        test["Death Predicted?"][i] = "Yes"
```
